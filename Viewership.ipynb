```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Here we explore how the vieweship varies throughout the series and which episodes are the most and least seen.

```python
kaggle=pd.read_csv('Data/Kaggle.csv')

# first take some information from the data 
kaggle.info()
```

```python
# Static
ax = sns.lineplot(kaggle.Season, kaggle.Viewership)
ax.set_xlabel('Season', fontsize=16)
ax.set_ylabel("Average Viewership per Episode", fontsize=16)
ax.set_title(f"Average Viewership per Episode Over All Seasons", fontsize=16)
sns.set(rc={'figure.figsize':(12,6)})
ax.get_figure().savefig('images/viewership_seasons.png')

# Interactive
# import chart_studio
# import chart_studio.plotly as py
# import plotly.express as px
# username='nklingen'
# api_key='C9j64BWfTfgzGrRDP5A3'
# chart_studio.tools.set_credentials_file(username=username,
#                                         api_key=api_key)

# fig = px.line(kaggle, x="Season", y="Viewership", title='Average Viewership per Episode Over All Seasons')
# fig.show()
```

The first thing we immediately notice is that The Office steadily increases in viewership until season 5. However, it then begins declining in seasons 6 and 7, even before Michael leaves at the end at season 7. In Seasons 8 and 9 where Michael is no longer a part of the cast, the viewership sharply declines until the series ends.

Moreover, we notice a wider variance in viewership in the first seasons. Over time, the variance tapers away, indicating that the show develops a "core" audience that watches all episodes. This is in contrast to Seasons 1 and 5, where we particularly notice a large variance in viewership between episodes of the same season

We see that the large variance in Season 1 can be explained by viewers who only watched the pilot. The pilot is the second most viewed episode of the Office, with 11.2 million views. Thus, this audience could have been drawn in by the initial hype, or were previous fans the the British version of the show, but were not hooked after the pilot. This is a very common pattern observed in pilot episodes.

```python
episodes_s1 = kaggle[kaggle.Season == 1].index + 1
viewership_s1 = kaggle[kaggle.Season == 1].Viewership
ax = sns.lineplot(episodes_s1, viewership_s1)
ax.set_xlabel('Episode', fontsize=16)
ax.set_ylabel("Viewership (Millions)", fontsize=16)
ax.set_title(f"Viewership per Episode in Season 1", fontsize=16)
sns.set(rc={'figure.figsize':(12,6)})
ax.get_figure().savefig('images/viewership_season1.png')
```

We see that the high variance in Season 5 can be explained by one episode in particular, that received a huge spike in viewership. Noteably, this episode, titled `Stress Relief` originally aired on February 1, 2009, immediately following NBC's broadcast of Super Bowl XLIII. Thus, they attracted a huge audience due to this prime time-slot. It had 22.91 million viewers - making it the most viewed episode of The Office. It didn't seem to actually gain them a larger audience in the longer run for the subsequent episodes in the series.

```python
episodes = [i+1 for i, x in enumerate(kaggle[kaggle.Season == 5].index)]
viewership_s2 = kaggle[kaggle.Season == 5].Viewership
ax = sns.lineplot(episodes, viewership_s2)
ax.set_xlabel('Episode', fontsize=16)
ax.set_ylabel("Viewership (Millions)", fontsize=16)
ax.set_title(f"Viewership per Episode in Season 5", fontsize=16)
sns.set(rc={'figure.figsize':(12,6)})
ax.get_figure().savefig('images/viewership_season5.png')
```
